## A Branch off the EVEREST pipeline
30 April 2019

The EVEREST pipeline is a promising tool for removing K2 systematics from faint objects' light curves. However, the use of cotrending basis vectors has a very strong possibility of also removing object variability. In this notebook, I will build off of EVEREST's Pixel Level Decorrelation(PLD) to create a simple pipeline that will remove systematics while preserving variability.

````python
from analysis import *
import everest

import logging
logging.getLogger("matplotlib").setLevel(logging.WARNING)
logging.getLogger("urllib3").setLevel(logging.WARNING)
````

For a test case we use a faint galaxy, requested by Ed Shaya in K2 Campaign 8, that we expect to have a flat light curve. This test case is used to develop our method. The raw light curves can be extracted using the ```lightkurve``` python package.

````python
# test object 1 (Ed Shaya) ch33
#epic1 = 220208401
epic1 = 220303559 #test 2
campaign=8


tpf = lk.search_targetpixelfile(epic1, campaign=campaign).download() # define tpf

# get raw lightcurve
lc = tpf.to_lightcurve(aperture_mask='all')
_, spurious_cad = lc.flatten().remove_outliers(return_mask=True) # calculate spurious cadences
lc_raw = lc.flux
cadno = np.arange(len(lc_raw)) # get an array to serve as our time/cadence measurement

# interpolate the spurious cadences
interped_vals = np.interp(cadno[spurious_cad], cadno[~spurious_cad], lc_raw[~spurious_cad])
# replace spurious cadence values with the interpolated values
lc_raw[spurious_cad] = interped_vals
````

````python
plt.plot(lc.time, lc_raw)
plt.xlabel("Time")
plt.ylabel("Raw flux")
````

Remember, we want this object to have a flat-looking light curve. 

````python
%%time
# apply everest
npld = everest.detrender.nPLD(epic1, season=campaign)
````

````python
# put flux/cadences into an array
# (there are 3852 cadences in a given 80 day campaign)
cad = np.arange(3852+1)
flux_pld = npld.flux

# turning indices found to be "bad" into a boolen mask to apply
mask = (np.isin(cad, np.concatenate([npld.nanmask, npld.badmask, npld.mask])))

# interpolate the spurious cadences
interped_vals = np.interp(cad[mask], cad[~mask], flux_pld[~mask])
# replace spurious cadence values with the interpolated values
flux_pld[mask] = interped_vals
````

````python
plt.plot(cad, flux_pld)
plt.xlabel("Cadence")
plt.ylabel("PLD flux")
````

We want to remove the rising trend that is in the PLD light curve, but rather than using cotrending basis vectors, we'll fit a linear slope and subtrect it. 

We're also going to remove the data for the first 3 days of the campaign (the "tail" in the beginning). 

````python
# 30 mintue intervals between cadences 
cutoff_day = 3*24*2
#cutoff = np.logical_and(cad>cutoff_day, cad<cad[-1]-5*cutoff_day)
cutoff = cad>cutoff_day
# finding linear fit 
m,b = np.polyfit(cad[cutoff], flux_pld[cutoff], 1)

# subtracting it
flux_corrected = flux_pld[cutoff] - (m*cad[cutoff])
````

````python
plt.plot(cad[cutoff], flux_corrected)
plt.xlabel("Cadence")
plt.ylabel("Corrected flux")
````

````python
# PSD before and after slope
fig,ax = plt.subplots(1,2, figsize=(18, 8))
plot_lc_PSD(cad/48.0, flux_pld,ax[0],ax[1], l='before slope subtraction', f=k2_freq, color='#eb2f06')
plot_lc_PSD(cad[cutoff]/48.0, flux_corrected,ax[0], ax[1],l='after slope subtraction', f=k2_freq, color='#6a89cc', alpha=0.9)
````

````python
# extract everest 
lc_everest = everest.Everest(epic2, season = 8)
````

````python
# put flux/cadences into an array
# (there are 3852 cadences in a given 80 day campaign)
cad = np.arange(3852+1)
flux_pld = lc_everest.flux

# turning indices found to be "bad" into a boolen mask to apply
mask = (np.isin(cad, np.concatenate([lc_everest.nanmask, lc_everest.badmask, lc_everest.mask])))

# interpolate the spurious cadences
interped_vals = np.interp(cad[mask], cad[~mask], flux_pld[~mask])
# replace spurious cadence values with the interpolated values
flux_pld[mask] = interped_vals
````

````python
plt.plot(cad, flux_pld)
plt.xlabel("Cadence")
plt.ylabel("PLD flux")
````

````python
# 30 mintue intervals between cadences 
cutoff_day = 3*(24*2)
#cutoff = np.logical_and(cad>cutoff_day, cad<cad[-1]-5*cutoff_day)
cutoff = cad>cutoff_day
# finding linear fit 
m,b = np.polyfit(cad[cutoff], flux_pld[cutoff], 1)

# subtracting it
flux_corrected = flux_pld[cutoff] - (m*cad[cutoff])
````

````python
plt.plot(cad[cutoff], flux_corrected)
plt.xlabel("Cadence")
plt.ylabel("Corrected flux")
````

````python
# PSD before and after slope
fig,ax = plt.subplots(1,2, figsize=(18, 8))
plot_lc_PSD(cad/48.0, flux_pld,ax[0],ax[1], l='before slope subtraction', f=k2_freq, color='#eb2f06')
plot_lc_PSD(cad[cutoff]/48.0, flux_corrected,ax[0], ax[1],l='after slope subtraction', f=k2_freq, color='#6a89cc', alpha=0.9)
````
