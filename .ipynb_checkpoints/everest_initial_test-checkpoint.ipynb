### Comparing Lighkurve PLD to EVERESTv2.0 PLD applied to K2 campaign 8 AGN
19 April 2019

In this notebook, I test using the everest python package to extract corrected light curves of AGN.

It's worth noting that I'm using the PLD flux that does not included de-trending using the covariance basis vectors. This is because we don't necessarily want the light curves to look flat! We want to study AGN variability.

```python
from analysis import *

from lightkurve.correctors import PLDCorrector
import everest 

import logging
logging.getLogger("matplotlib").setLevel(logging.WARNING)
logging.getLogger("urllib3").setLevel(logging.WARNING)
```

```python
# Checking which channels have most of our AGN on them?
agn = pd.read_csv('./object_keys/c8_agn_w_redshift.csv')
ch = np.array(agn['CHANNEL']).astype(int)

# sort
s = np.argsort(np.bincount(ch))

# print sorted by occurence (low->high)
print(s)
```

---
## AGN \# 1
Has a notable m-shape like some of the Ed shaya galaxies

```python
epic = 220179771
campaign = 8 
```

```python
lc_everest = everest.Everest(epic)
#everest.Search(epic)
```

```python
tpf = lk.search_targetpixelfile(epic, campaign=campaign).download() # define tpf

# get raw lightcurve
lc = tpf.to_lightcurve(aperture_mask='all')
_, spurious_cadences = lc.flatten().remove_outliers(return_mask=True) # remove spurious cadences
lc_raw = lc[~spurious_cadences]
```

```python
# get PLD corrected (lightkurve)
pld = PLDCorrector(tpf)# instantiate pld corrector
lc_pld = pld.correct()# correct with 2nd order pld
lc_pld = lc_pld[~spurious_cadences]# remove spurious cadences
```

```python
#lc_raw.plot(marker='.', ls=' ')
plt.figure(figsize=(13.5,6))
plt.plot(lc_raw.time, lc_raw.flux, '.', alpha = 0.8)
plt.title("Raw Flux")

# lightkurve PLD corrected
#lc_pld.plot(marker='.', ls=' ')

plt.figure(figsize=(13.5,6))
plt.plot(lc_pld.time, lc_pld.flux, '.', alpha = 0.8)
plt.title("Lightcurve PLD")
```

```python
# everest pipeline
_ = lc_everest.plot(plot_cbv=False)
```

The bottom light curve can be called with the following code:

```python
# define a mask that will remove [nanmask], [basmask], and spurious cadences
mask = np.isin(np.arange(3852+1), np.concatenate([lc_everest.nanmask, lc_everest.badmask, lc_everest.mask]))

fig, ax = plt.subplots(figsize=(12,6))
ax.plot(lc_everest.time[~mask], lc_everest.flux[~mask])
```

However, when we try to plot this for ourselves, it looks like a hot mess.

```python
plt.plot(lc_everest.time, lc_everest.flux,'.')
```

I've dug into the EVEREST code on github to create a (negation) boolean mask to apply to the time and flux arrays, and get the light curve in black, on the bottom EVEREST plots.

```python
mask = np.isin(np.arange(3852+1), np.concatenate([lc_everest.nanmask, lc_everest.badmask])) #, lc_everest.mask]))

sc = np.isin(np.arange(3852+1), lc_everest.mask)
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(lc_everest.time[~mask], lc_everest.flux[~mask], 'o')
ax.plot(lc_everest.time[sc], lc_everest.flux[sc],'r.')
ax.plot(lc_everest.time[mask], lc_everest.flux[mask], 'g.')
```

---
The PSDs for the above first and third objects look mostly flat. Couldn't calculate the PSD for the middle one for some reason...

Some other AGN to look at: 220250435 (PSD looks right), 220343599 (PSD looks flat), 220171966 (PSD looks flat), 220452157 (PSD looks mostly flat)

```python
epics = [220179771, 220444932, 220226374,  220250435, 220343599, 220171966, 220452157]
campaign = 8

for epic in epics:
    tpf = lk.search_targetpixelfile(epic, campaign=campaign).download() # define tpf
    channel = tpf.channel
    kepmag = tpf.hdu[0].header['KEPMAG']# getting magnitude
    
    # get raw lightcurve
    lc = tpf.to_lightcurve(aperture_mask='all')
    _, spurious_cadences = lc.flatten().remove_outliers(return_mask=True) # remove spurious cadences
    lc_raw = lc[~spurious_cadences]

    # get PLD corrected (everest 2.0)
    lc_everest = everest.Everest(epic)
    mask = np.isin(np.arange(3852+1), np.concatenate([lc_everest.nanmask, lc_everest.badmask,  lc_everest.mask]))
    lc_eve_time = lc_everest.time[~mask]
    lc_eve_flux = lc_everest.flux[~mask]
    
    # plot them
    fig, ax = plt.subplots(1,2, figsize=(18,6))
    plot_lc_PSD(lc_raw.time, lc_raw.flux, ax[0], ax[1], l="Raw", f=k2_freq, color='k', alpha=1.0)
    plot_lc_PSD( lc_eve_time, lc_eve_flux, ax[0], ax[1], l="PLD", f=k2_freq, color='b', alpha=0.9)
    fig.suptitle("EPIC %s, K2 Campaign %s Ch %s (Mag = %s)"%(epic,campaign,channel,kepmag), fontsize=16)
    plt.savefig("/home/rachel/Research/K2/submit/c8_%s.png"%epic)
```
